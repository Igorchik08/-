Ознайомився з циклами, написав декілька прикладів їх роботи:

код програми
letters = ["a", "b", "c"]
print("list of letters:")
for i in range(len(letters)):
    print(f"{i+1}. {letters[i]}")

i = 0
while letters[i] != "c":
    print(f"{i}. {letters[i]}")
    i += 1
результат виконання
list of letters:
1. a
2. b
3. c

0. a
1. b
Ознайомився з принципом роботи розгалуджень, виконав декілька прикладів їх роботи:

код програми

print(f"name = {name}" if "name" in locals() else "variable `name` doesnt exist")
результат виконання
b = 1
variable `name` doesnt exist
Приклад роботи конструкції try->except->finallyдля виявлення можливих помилок у коді:

код програми
negNum = -100
try: 
    print(f"square root of {negNum} equals", math.sqrt(negNum))
except Exception as e:
    print("ERROR:", e)
finally:
    print("(this math operation cannot be performed)")
результат виконання
ERROR: math domain error
(this math operation cannot be performed)
